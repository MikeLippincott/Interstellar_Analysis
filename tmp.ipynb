```python
import pathlib

import numpy as np
import pandas as pd
```

```python
# set path
path = pathlib.Path("all_runs.txt")

keyword = "cell_type"  # replace with your keyword
groups = []
current_group = []

with open(path, "r") as file:
    for line in file:
        if keyword in line:
            if current_group:  # if current_group is not empty, add it to groups
                groups.append(current_group)
                current_group = []  # start a new group
        current_group.append(line)

# Add the last group if it's not empty
if current_group:
    groups.append(current_group)

# Now, 'groups' is a list of groups, where each group is a list of lines.
# You can print the groups to check the result:
for i, group in enumerate(groups):
    print(f"Group {i+1}:")
    for line in group:
        print(line)
```

```python
# remove lines that contain conda
for group in groups:
    group[:] = [line for line in group if "conda" not in line]
```

```python
# return groups that contain "slurmstepd: error: ***"
error_groups = [group for group in groups if any("slurm" in line for line in group)]
print(error_groups)
len(error_groups)
# # parse the groups for the first line
error_groups = [group[0] for group in error_groups]
error_groups
```
