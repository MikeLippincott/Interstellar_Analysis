```r
# import csv file
df <- read.csv(
    "./results/single_class/PBMC/aggregated_with_nomic/regression/DMSO_0.100_DMSO_0.025__LPS_100.000_DMSO_0.025/model_stats.csv", 
    )
head(df,2)


enet_cp_fig <- "./figures/single_class/PBMC/aggregated_with_nomic/DMSO_0.100_DMSO_0.025__LPS_100.000_DMSO_0.025/enet_cp.pdf"
```

```r
unique(df$shuffled)
```

```r
# set the plot size
options(repr.plot.width=10, repr.plot.height=8)
# plot the data
global_prediction_trend <- (
    ggplot(df, aes(x=actual, y=predicted, col=shuffled)) 
    + geom_point() 
    + geom_smooth(method=lm, se=TRUE,) 
    + labs(x="Actual", y="Predicted") 
    + theme_bw()
    # add y=x line
    + geom_abline(intercept = 0, slope = 1, linetype="dashed", color="black")
)
global_prediction_trend
```

```r
pdf(file=enet_cp_fig)
# set plot size
options(repr.plot.width=6, repr.plot.height=8)
# facet by secrete 
for (i in 1:length(unique(df$secrete))){
    sub_df <- df[df$secrete == (unique(df$secrete)[i]),]
    print(unique(sub_df$secrete))
    p <- (
        ggplot(sub_df, aes(x=actual, y=predicted, col=shuffled)) 
        + geom_point() 
        + geom_smooth(method=lm, se=TRUE) 
        + labs(x="Actual", y="Predicted") 
        + theme_bw() 
        + ggtitle(unique(df$secrete)[i])
        + ylim(0, 1)
        + xlim(0, 1)

    )
    
    plot(p)
}
dev.off()
```
